### Выделение улиц в столбце address
Выделите в отдельный столбец информацию об улице из столбца address.

```python
# Исключим все поселения:
data_address = data[data.apply(lambda x: False if 'поселение' in x['address'] else True, axis=1)].copy()
```

```python
# избавимся от "города Москва, ":
regex_pat = re.compile(r'город Москва, ', flags=re.IGNORECASE)
data_address['address'] = data['address'].str.replace(regex_pat, '')

# Добавим ту же самую предобработку данных с приведением к нижнему регистру и заменой кавычек:
data_address['object_name'] = data_address['object_name'].str.lower().str.replace('«', '').str.replace('»', '')

# Напишем функцию которая поможет 
def splt_comma(str_in, n_from, n_to):
    if n_to is None:
        return str(str_in.split(',')[n_from])
    
    if n_to is not None:
        return str_in.split(',')[n_from : n_to]
    
def street(string):
    if (('город' in string) or ('поселок' in string)) and not ('городская' in string):
        if len(splt_comma(string,-3,-1)) > 1: 
            return ','.join(splt_comma(string,-3,-1))
        else: 
            return splt_comma(string,0,None)
    if 'поселение' in string: 
        return splt_comma(string,-2,-1)
    else: 
        return splt_comma(string,0,None)

# Создадим столбец street:
data_address['street'] = data_address['address'].apply(street)
```

### Топ-10 улиц
Постройте график топ-10 улиц по количеству объектов общественного питания. Воспользуйтесь внешней информацией и ответьте на вопрос — в каких районах Москвы находятся эти улицы? 

```python
# Соберём топ-10 улиц и отсортируем их по количеству объектов, при этом удалим все поселения и отдельные города, 
# так сравнивать небольшой город с улицей некорректно - масштаб этого несопоставим:

data_streets = data_address[data_address.apply(
    lambda x: False if ('город' in x['street']) or ('посёлок' in x['street']) else True, axis=1)].copy()

data_streets = data_streets.groupby('street', as_index=False).agg(
    {'id':'count'}).rename(columns={'id':'num_of_rests'})
top_10 = data_streets.sort_values(by='num_of_rests', ascending=False).head(10)
top_10
```

```python
address_registry = pd.read_csv('mos_ru_dataset.csv', sep=';', comment='#')
# Переименуем имеющиеся колонки:
address_registry = address_registry.rename(columns={
    'P1':'city', 'P5':'district', 'P7':'street', 'ADM_AREA': 'adm_area'})
# Сгруппируем по названию улицы:
address_registry = address_registry.groupby('street', as_index=False).first()
# и выведем его:
address_registry
```

```python
# Присоединим районы и округа к топ-10 улиц:
top_10_joined = top_10.merge(address_registry, on='street').drop(columns='city')
top_10_joined
```

```python
top_10_grouped = top_10_joined.groupby('adm_area').agg({'num_of_rests':'sum'}).sort_values('num_of_rests', ascending=False).copy()
top_10_grouped
```

```python
# Много это или мало? Добавим другие показатели:
data_population = pd.read_csv('moscow_population_density.csv', header=1)
data_population = data_population.rename(
    columns={"Адми-\nнистра-\nтивный\nокруг":'adm_area', 
             "Пло-\nщадь,\n[6][7]\nкм²":'area', 'Насе-\nление\n2019[8],\nчел.':'population'})

# Сделаем небольшую предобработку данных: 
data_population_clean = data_population[['adm_area','area','population']].copy()
data_population_clean['population'] = data_population_clean.apply(
    lambda x: re.sub(r'[↗|↘| ]', '', x['population']), axis=1).copy().astype('int')
data_population_clean['area'] = data_population_clean.apply(
    lambda x: re.sub(r'[,]', '.', x['area']), axis=1).copy().astype('float').copy()

# Добавим столбец по плотности населения:
data_population_clean_grouped = data_population_clean.groupby('adm_area', as_index=False).sum()
data_population_clean_grouped['population_density'] = data_population_clean_grouped['population']\
                                                            /data_population_clean_grouped['area']

# Переделаем названия чтобы они соответствовали сокращённым:
adm_district_list = address_registry['adm_area'].unique()
data_population_clean_grouped.sort_values('population_density')

# Добавим соответствия для длинных имён (к сожалению, руками):
def rename_adm_district(short_name):
    if short_name['adm_area'] == 'ТАО': return 'Троицкий административный округ'  
    if short_name['adm_area'] == 'НАО': return 'Новомосковский административный округ'
    if short_name['adm_area'] == 'ЗелАО': return 'Зеленоградский административный округ'
    if short_name['adm_area'] == 'ЗАО': return 'Западный административный округ'
    if short_name['adm_area'] == 'САО': return 'Северный административный округ'
    if short_name['adm_area'] == 'СВАО': return 'Северо-Восточный административный округ'
    if short_name['adm_area'] == 'СЗАО': return 'Северо-Западный административный округ'
    if short_name['adm_area'] == 'ЦАО': return 'Центральный административный округ'
    if short_name['adm_area'] == 'ЮАО': return 'Южный административный округ'
    if short_name['adm_area'] == 'ЮВАО': return 'Юго-Восточный административный округ'
    if short_name['adm_area'] == 'ЮЗАО': return 'Юго-Западный административный округ'
    if short_name['adm_area'] == 'ВАО': return 'Восточный административный округ'
    
# Применим функцию:
data_population_clean_grouped['adm_area'] = data_population_clean_grouped.apply(
                                                            rename_adm_district, axis=1)
data_population_clean_grouped
```

```python
# Соберём датафреймы с плотностью населения:
top_10_joined_final = top_10_joined.merge(
    data_population_clean_grouped, on='adm_area').round(1).sort_values('num_of_rests', ascending=False)
top_10_joined_final
```
